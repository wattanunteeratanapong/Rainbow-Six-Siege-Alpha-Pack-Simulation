```python
import random
import matplotlib.pyplot as plt
```

```python
match_list = []
match_data = []
correlation = {}

def chance(percent):
    random_number = random.random()
    return random_number < (percent / 100)

def alphapack():
    match_list = [random.randint(0, 1) for i in range(40)]
    percent = 0
    matchs = 0
    for match in match_list:
        win_count = 0
        if match == 1:
            percent = percent + 2
            matchs = matchs + 1
            win_count = win_count + 1
            if chance(percent):
                match_data.append(matchs)
                correlation[win_count] = correlation.get(win_count, 0) + 1
                return
        elif match == 0:
            percent = percent + 1.5
            matchs = matchs + 1

def simulation(generate_times):
    for i in range(generate_times):
        alphapack()
```

```python
simulation(10000000)
```

```python
fig, axs = plt.subplots(1, 1, figsize=(25, 13))

axs.hist(match_data, bins=range(1, 40), edgecolor='black', density=False) 
axs.set_xticks(range(1, 40)) 
axs.set_xlabel('Number of Matches')
axs.set_ylabel('Frequency')
axs.set_title('Distribution of Matches to Obtain Alpha Pack (Standard)', fontsize=17)
axs.grid(True)

plt.show()
```
